```python
# بسم الله الرحمن الرحيم
# (وما توفيقى الا بالله)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv(r"C:\Users\moham\Downloads\workspace (1)\workspace\production_data.csv")

df.head()
```

Task 1:Cleaning Steps

```python
## Parsing Date

df['production_date']= pd.to_datetime(df['production_date'])

## raw_material_supplier COl


df['raw_material_supplier'] = df['raw_material_supplier'].replace(1, 'national_supplier')
df['raw_material_supplier'] = df['raw_material_supplier'].replace(2, 'international_supplier')
df['raw_material_supplier'] = df['raw_material_supplier'].astype(str)

## pigment_type Col

df['pigment_type'] = df['pigment_type'].replace('Type_C','type_c') 

## pigment_quantity

df['pigment_quantity'] = df['pigment_quantity'].astype(int)

# pigment_quantity

df['pigment_quantity'] = df['pigment_quantity'].astype(int)

## mixing_time and filling NAN and rounding

df['mixing_time'].fillna(df['mixing_time'].mean(),inplace=True)
df['mixing_time'] = df['mixing_time'].round(2)

## mixing_speed Col 

df['mixing_speed'] = df['mixing_speed'].replace("-","Not Specified")

## product_quality_score

df['product_quality_score'] = df['product_quality_score'].round(2)


clean_data = df

clean_data.head()
```

Task 2 : Calculate the average product_quality_score and pigment_quantity grouped by raw_material_supplier

```python
aggregated_data = clean_data.groupby('raw_material_supplier').agg(
    avg_product_quality_score = ('product_quality_score','mean'),
    avg_pigment_quantity = ('pigment_quantity','mean')
).round(2).reset_index()


aggregated_data
```

Task 3 : Identify the average product_quality_score for batches with a raw_material_supplier of 2 and a pigment_quantity greater than 35 kg.

1) Your output should be a DataFrame named pigment_data.
2) It should consist of a 1-row DataFrame with 3 columns: raw_material_supplier, pigment_quantity, and avg_product_quality_score.
3) Your answers should be rounded to 2 decimal places where appropriate.

```python
# Write your answer to Task 3 here

filtered = clean_data[(clean_data['raw_material_supplier']=='international_supplier')&(clean_data['pigment_quantity']>35)]

pigment_data = filtered.groupby(['raw_material_supplier', 'pigment_quantity'])['product_quality_score'] \
    .mean().round(2).reset_index() \
    .rename(columns={'product_quality_score': 'avg_product_quality_score'})

pigment_data
```

Task 4 : 
1) Calculate the mean and standard deviation for the columns pigment_quantity and product_quality_score as: product_quality_score_mean, product_quality_score_sd, pigment_quantity_mean, pigment_quantity_sd.
2) Calculate the Pearson correlation coefficient between pigment_quantity and product_quality_score as: corr_coef
Your output should be a DataFrame named product_quality.

```python
# Write your answer to Task 4 here

product_quality_score_mean = clean_data['product_quality_score'].mean().round(2)
product_quality_score_sd = clean_data['product_quality_score'].std().round(2)
pigment_quantity_mean = clean_data['pigment_quantity'].mean().round(2)
pigment_quantity_sd = clean_data['pigment_quantity'].std().round(2)
corr_coef = clean_data['pigment_quantity'].corr(clean_data['product_quality_score']).round(2)

dict = {
    'product_quality_score_mean': [product_quality_score_mean],
    'product_quality_score_sd' : [product_quality_score_sd],
    'pigment_quantity_mean' : [pigment_quantity_mean],
    'pigment_quantity_sd' : [pigment_quantity_sd],
    'corr_coef' : corr_coef
}

product_quality = pd.DataFrame(dict)

product_quality

```
